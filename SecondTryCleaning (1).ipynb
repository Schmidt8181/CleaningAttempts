```python
# Adjust column names and move on to clean 1

welcome_2013.columns = ['Publisher', 'Journal', 'Article title', 'Cost'] 
# This change is inplace, don't know if i want that yet.
# welcome_2013 = welcome_2013.drop('PMID_PMCID', axis=1)
welcome_2013 = welcome_2013.drop('Publisher', axis=1)

# drop unneeded columns. The PMCID number is a reference number for the individual articles,
# and the Publisher is irrelivent to the questions we're answering. 
# The columns were dropped in place
```

```python
# for first cleaning step, remove white space and lower the case of the entire frame,
# and group by journal name, with counts sorted by the cost column
# newclean2 = ()

def clean_one(clean_var1):
    newclean1 = clean_var1.apply(lambda x: x.astype(str).str.lower())
    newclean2 = newclean1.apply(lambda x: x.str.strip())
    return(newclean2.groupby('Journal').count().sort_values('Cost', ascending=False, kind='mergesort'))
    
clean_one(welcome_2013)
```

```python
# for cleaning step two, drop null/empty values, and characters causing trouble in the Journal column
def clean_two(clean2_var1):
    clean2_var2 = clean2_var1.apply(lambda x: x.dropna())
    clean2_var3 = clean2_var2.apply(lambda x: x.str.strip('the', axis=0, inplace=False))
    print(clean2_var3)
    
clean_two(welcome_2013)
```
